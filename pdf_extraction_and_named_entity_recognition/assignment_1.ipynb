## Reading blocks using pymupdf

```python
new_text = page.get_text("dict")["blocks"]
revent_data_new = []
for i in range(len(new_text)):
    if ( (new_text[i]['type'] == 0)  & (new_text[i]['bbox'][0] == 51.02360153198242 or new_text[i]['bbox'][0] == 222.99209594726562 or\
                                new_text[i]['bbox'][0] == 394.9606018066406 )): ### excluding image blocks  
        print(new_text[i]['bbox'][0],new_text[i]['bbox'][1],new_text[i]['bbox'][2],new_text[i]['bbox'][3])
        revent_data_new.append(new_text[i])
```

## Saving text and font size to keep track of headings.

```python
text_blocks = {}
x = 0
text_blocks[0] = ''
for i in range(len(revent_data_new)):
    for j in range(len(revent_data_new[i]['lines'])):
        y = text_blocks[x]
        text_blocks[x] = y + revent_data_new[i]['lines'][j]['spans'][0]['text'].replace(u'\xa0', u' ')
        if revent_data_new[i]['lines'][j]['spans'][0]['font'] == 'Roboto-Bold':
            text_blocks[x] = [text_blocks[x],'Roboto-Bold']
            x = x+1
            text_blocks[x] = ''
    if text_blocks[x][-1] == '.' or text_blocks[x][-2] == '.' :
        text_blocks[x] = [text_blocks[x],'Roboto-Light']
        x = x+1
        text_blocks[x] = ''

text_blocks[x] = [text_blocks[x],'Roboto-Light']
```

```python
text_blocks  ## need to make two paragraph one if they dont end by a full stop
```

## Writting the data into csv file

```python
import csv
with open('assgnment_1.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in text_blocks.items():
        writer.writerow([key, value[0],value[1]])
```
